# 2024 Earthquake Data Analysis Project

## Overview

This project uses US earthquake data to perform exploratory data analysis (EDA), visualize earthquake patterns, and investigate seismic events across the United States. The data is sourced from the U.S. Geological Survey (USGS) Earthquake Database, which provides comprehensive, real-time information about seismic activity.

## Key Data Features:

The dataset contains the following columns:

* time: Date and time of the earthquake
* latitude: Latitude of the earthquake epicenter
* longitude: Longitude of the earthquake epicenter
* depth: Depth of the earthquake (in kilometers)
* mag: Magnitude of the earthquake
* magType: Type of magnitude measurement (e.g., Moment Magnitude)
* nst: Number of stations reporting the event
* gap: Azimuthal gap (in degrees)
* dmin: Minimum distance to the earthquake
* rms: Root mean square value of amplitude spectra
* net: Network code
* id: Unique identifier for each earthquake
* updated: Last update timestamp for the earthquake data
* place: Location of the earthquake (e.g., a specific city or region)
* type: Type of earthquake (e.g., "earthquake", "explosion")
* horizontalError: Horizontal location error (in km)
* depthError: Error in depth measurement (in km)
* magError: Error in magnitude measurement
* magNst: Number of stations used to determine the magnitude
* status: Status of the earthquake report (e.g., "Reviewed", "Automatic")
* locationSource: Source of location data
* magSource: Source of magnitude data

# Project Goals

The primary goal of this project is to analyze and visualize earthquake data to understand patterns in seismic activity in the United States. By exploring various features of the dataset, we aim to answer questions such as:

1. What regions of the U.S. experience the most earthquakes?
2. Is there a relationship between earthquake magnitude and depth?
3. Is there a relationship between the number of earthquakes and the month of the year?

# Data Filtering and Focus
For this analysis, only earthquakes with a magnitude â‰¥ 2 have been included in the dataset. This threshold has been chosen because earthquakes with a magnitude of 2 or greater are generally sensed by people, especially those living near the epicenter.
Smaller earthquakes (mag < 2) typically do not have a significant impact on humans and are often not felt by people, although they can still be detected by seismographs.
In addition, only Conterminous U.S.([24.6, 50] Latitude, [-125, -65] Longitude) events were considered for the Geographic Region, as provided by USGS Earthquake.
